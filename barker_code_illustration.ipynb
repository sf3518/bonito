```python
import numpy as np
import copy
import matplotlib.pyplot as plt
from scipy import signal
```

```python
barker7 = np.array([1, 1, 1, -1, -1, 1, -1])
xs = np.linspace(0, 7, 7, endpoint=False)
plt.plot(xs, barker7, 'go-')
```

```python
# Dot product with itself is 7
dot_prod = barker7.dot(barker7.T)
print(dot_prod)
```

```python
# Correlate barker code with itself
cor = np.correlate(barker7, barker7, 'full')
print(cor)
plt.plot(cor)
```

```python
# Resample 1 digital signal sample into 100
barker_resample = signal.resample(barker7, 700)
xs_new = np.linspace(0, 7, 700, endpoint=False)
# Plot the resampled barker code
plt.plot(xs, barker7, 'go-', xs_new, barker_resample, '.-', 7, barker_resample[0], 'ro')
```

```python
# Add 10 offset to barker code
barker_offset = [barker_resample[i] for i in range(10, 700, 100)]
print(barker_offset)
```

```python
plt.plot(xs, barker_offset, 'go-')
```

```python
# Plot autocorrelation for barker code with offset
cor_offset = np.correlate(barker7, barker_offset, 'full')
print(cor_offset)
plt.plot(cor_offset)
```

```python
# Plot autocorrelation for offset 10, 20, 30, 40, 50 in one figure

for i in range(6):
  barker_offset = [barker_resample[j] for j in range(10 * i, 700, 100)]
  cor_offset = np.correlate(barker7, barker_offset, 'full')
  plt.plot(cor_offset)
```
